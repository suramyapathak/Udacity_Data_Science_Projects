```python
cat_0=re_check_data[(re_check_data.viewed==0) & (re_check_data.completed==0)]
cat_1=re_check_data[(re_check_data.viewed==1) & (re_check_data.completed==0)]
cat_2=re_check_data[(re_check_data.viewed==0) & (re_check_data.completed==1)]
cat_3=re_check_data[(re_check_data.viewed==1) & (re_check_data.completed==1)]

bogo=re_check_data[(re_check_data.bogo==1)]
discount=re_check_data[(re_check_data.discount==1)]

bogo_0=cat_0[(cat_0.bogo==1)]
bogo_1=cat_1[(cat_1.bogo==1)]
bogo_2=cat_2[(cat_2.bogo==1)]
bogo_3=cat_3[(cat_3.bogo==1)]

discount_0=cat_0[(cat_0.discount==1)]
discount_1=cat_1[(cat_1.discount==1)]
discount_2=cat_2[(cat_2.discount==1)]
discount_3=cat_3[(cat_3.discount==1)]

cat_0.reset_index(drop=True, inplace=True)
cat_1.reset_index(drop=True, inplace=True)
cat_2.reset_index(drop=True, inplace=True)
cat_3.reset_index(drop=True, inplace=True)
bogo.reset_index(drop=True, inplace=True)
discount.reset_index(drop=True, inplace=True)

discount_0.reset_index(drop=True, inplace=True)
discount_1.reset_index(drop=True, inplace=True)
discount_2.reset_index(drop=True, inplace=True)
discount_3.reset_index(drop=True, inplace=True)

bogo_0.reset_index(drop=True, inplace=True)
bogo_1.reset_index(drop=True, inplace=True)
bogo_2.reset_index(drop=True, inplace=True)
bogo_3.reset_index(drop=True, inplace=True)
```

```python
X=bogo.drop(['success','customer_id', 'received', 'completed','bogo'], axis=1)
y=bogo.success

X_train, X_test, y_train, y_test = train_test_split(X,y, 
                                                        test_size=0.20, 
                                                        random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=42)
   


model = XGBClassifier(learning_rate = 0.1,num_class=4,max_depth = 7,min_child_weight = 5,objective = 'multi:softmax',seed = 42,gamma = 0.1,silent = True)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
X_test=scaler.transform(X_test)
```

```python
eval_set=[(X_val,y_val)]
```

```python
 model = XGBClassifier(learning_rate = 0.1,\
                                  num_class=4,\
                                      max_depth = 7,\
                                      min_child_weight = 5,\
                                      objective = 'multi:softmax',\
                                      seed = 42,\
                                      gamma = 0.1,\
                                      silent = True)
```

```python
model.fit(X_train, y_train,eval_set=eval_set,eval_metric="mlogloss", verbose=True)
```

```python
# create confusion matrix heat map 
valid_pred = model.predict(X_test)
    
# print the accuracy score for the validation data    
accuracy = classification_report(y_test, valid_pred)
print(accuracy, 'accuracy')
    
cm = confusion_matrix(y_test, valid_pred)

fig, ax= plt.subplots(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='g', ax = ax); #annot=True to annotate cells

# labels, title and ticks
ax.set_xlabel('Predicted labels');
ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 


# features importance 
plot_importance(model)
```

```python
def XGBClassifierGrid(X_train, y_train):
    clf= XGBClassifier(num_class=4, objective = 'multi:softmax')
    param_grid = {
        'n_estimators': [25, 50, 100, 200, 500],
        'max_depth': [3, 5, 8 ,10],
        'gamma': [0.01, 0.1, 0.5, 1, 1.5, 2, 6],
        'subsample':[0.5,  0.8,  1.0],
        'colsample_bytree': [0.3, 0.5, 0.8, 1.0],
        'learning_rate': [0.01, 0.05, 0.1, 0.2], 
        'min_child_weight': [0.5, 1.0, 3.0, 5.0, 10.0],
        'silent': [False]
    }
    fit_params = {
        "verbose":True,
       # "eval_set":eval_set
    }
    rs_clf = RandomizedSearchCV(clf, param_grid, n_iter=20,
                           scoring='f1_weighted',  random_state=42)
    print("Randomized search..")
    rs_clf.fit(X_train, y_train)
    best_score = rs_clf.best_score_
    best_params = rs_clf.best_params_
    print("Best params: ")
    for param_name in sorted(best_params.keys()):
        print('%s: %r' % (param_name, best_params[param_name]))
        model = rs_clf.best_estimator_
    return model
```
